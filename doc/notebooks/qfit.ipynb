```python
import numpy as np
```

```python
custom_fit = qfit.QFIT()
```

```python
coordinate = np.linspace(0,10,200)
data = np.linspace(1,41,200)
```

```python
custom_fit.load(coordinate=coordinate,data=data)
```

```python
def f_custom(coord, a, b):
    return a+b*coord   #line
```

```python
custom_fit.fit_function = f_custom
custom_fit.p0 = [1,5]
custom_fit.fit()
```
